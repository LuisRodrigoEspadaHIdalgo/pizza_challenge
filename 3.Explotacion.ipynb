## ¿Qué días y horas tendemos a estar más ocupados?

Pensando en cantidades de pizzas preparadas

```
%sql
SELECT    SUM(quantity) AS cantidad_de_pizzas,day_week
FROM      fact_orders
LEFT JOIN dim_time
ON        dim_time_id = dim_time.id
GROUP BY  day_week
ORDER BY  cantidad_de_pizzas DESC
LIMIT 20

```

Pensando en cantidades de ordenes recibidas

```
%sql
SELECT    COUNT(DISTINCT fact_orders.order_id) AS cantidad_de_ordenes,day_week
FROM      fact_orders
LEFT JOIN dim_time
ON        dim_time_id = dim_time.id
GROUP BY  day_week
ORDER BY  cantidad_de_ordenes DESC
LIMIT 20
```

Vemos que los días mas ocupados son Jueves, Viernes, Sabado y Domingo.

Mismo razonamiento para las horas...

Pensando en cantidades de pizzas preparadas

```
%sql
SELECT    SUM(quantity) AS cantidad_de_pizzas,hour
FROM      fact_orders
LEFT JOIN dim_time
ON        dim_time_id = dim_time.id
GROUP BY  hour
ORDER BY  cantidad_de_pizzas DESC
LIMIT 20
```

Pensando en cantidades de ordenes recibidas

```
%sql
SELECT    COUNT(DISTINCT fact_orders.order_id) AS cantidad_de_ordenes,hour
FROM      fact_orders
LEFT JOIN dim_time
ON        dim_time_id = dim_time.id
GROUP BY  hour
ORDER BY  cantidad_de_ordenes DESC
LIMIT 20
```

Horas mas ocupadas: 12,13,18,17,19 horas.

## ¿Cuántas pizzas estamos haciendo durante los períodos pico?

```
%sql
SELECT    SUM(quantity) AS cantidad_de_pizzas,day_week
FROM      fact_orders
LEFT JOIN dim_time
ON        dim_time_id = dim_time.id
GROUP BY  day_week
ORDER BY  cantidad_de_pizzas DESC
LIMIT 4
```

```

%sql
SELECT    SUM(quantity) AS cantidad_de_pizzas,hour
FROM      fact_orders
LEFT JOIN dim_time
ON        dim_time_id = dim_time.id
GROUP BY  hour
ORDER BY  cantidad_de_pizzas DESC
LIMIT 4
```

##¿Cuáles son nuestras pizzas más vendidas y peores?

```
%sql

  SELECT    SUM(quantity) AS cantidad_de_pizzas,dim_pizzas.id,name,size,category
  FROM      fact_orders
  LEFT JOIN dim_pizzas
  ON        dim_pizzas_id = dim_pizzas.id
  GROUP BY  dim_pizzas.id,name,size,category
  ORDER BY  cantidad_de_pizzas DESC
```

* Tenemos que las pizzas mas pedidas son aquellas con id = 6,86,23,26
* Las menos pedidas con id = 93,29,16,10

```
%sql

  SELECT    SUM(quantity) AS cantidad_de_pizzas,dim_pizzas.id,name,size,category
  FROM      fact_orders
  LEFT JOIN dim_pizzas
  ON        dim_pizzas_id = dim_pizzas.id
  GROUP BY  dim_pizzas.id,name,size,category
  ORDER BY  cantidad_de_pizzas DESC
  LIMIT 4
```

```
%sql

  SELECT    SUM(quantity) AS cantidad_de_pizzas,dim_pizzas.id,name,size,category
  FROM      fact_orders
  LEFT JOIN dim_pizzas
  ON        dim_pizzas_id = dim_pizzas.id
  GROUP BY  dim_pizzas.id,name,size,category
  ORDER BY  cantidad_de_pizzas ASC
  LIMIT 4
```
